```python
# Number of deals in each season

sns.set_style("whitegrid")
sns.countplot(x="Season Number", hue="Got Deal", data=data)
plt.figure(figsize=(15, 15))
plt.show()


#Season 9, 10, 11 had more deal as compared to rejections
#Season 1,3,5 had more rejection as compared to other
#Starting season 6 there are much more accepted deals as compared to rejection
```

```python
#Gender and getting deals
data_deals = data[data['Got Deal'] == 1]

# calculate the total number of pitches that got the deal for each gender category
total_female_deals = len(data_deals[data_deals['Pitchers Gender'] == 'Female'])
total_male_deals = len(data_deals[data_deals['Pitchers Gender'] == 'Male'])
total_mixed_deals = len(data_deals[data_deals['Pitchers Gender'] == 'Mixed Team'])

# create a bar chart of the total number of pitches that got the deal for each gender category
plt.bar(['Female', 'Male', 'Mixed Team'], [total_female_deals, total_male_deals, total_mixed_deals])

# set the title and axis labels
plt.title("Total Pitches That Got the Deal by Gender")
plt.xlabel("Gender")
plt.ylabel("Number of Pitches")

# show the plot
plt.show()
```

```python
#Gender wise deals accepted and rejected
import seaborn as sns
import matplotlib.pyplot as plt

# filter for pitches where got deal = 1
data_deals = data[data['Got Deal'] == 1]

# create a countplot of the gender and got deal columns
sns.countplot(x='Pitchers Gender', hue='Got Deal', data=data, palette=['#3182bd', '#e6550d'])

# set the title and axis labels
plt.title("Effect of Gender on Deal Outcomes")
plt.xlabel("Gender")
plt.ylabel("Number of Pitches")

# show the plot
plt.show()
```

```python
#Total Accepted - Rejected Gender wise

# group the data by gender and got deal status, and count the number of pitches in each group
grouped_data = data.groupby(['Pitchers Gender', 'Got Deal'])['Pitch Number'].count()

# calculate the difference between the count of pitches with got deal = 1 and got deal = 0 for each gender
female_diff = grouped_data['Female', 1] - grouped_data['Female', 0]
male_diff = grouped_data['Male', 1] - grouped_data['Male', 0]
mixed_diff = grouped_data['Mixed Team', 1] - grouped_data['Mixed Team', 0]

# print the differences for each gender
print(f"Female pitch difference: {female_diff}")
print(f"Male pitch difference: {male_diff}")
print(f"Mixed pitch difference: {mixed_diff}")
```

```python
#Which season had the most female pitches?

# filter the data for pitches thrown by female pitchers
female_pitches = data[data['Pitchers Gender'] == 'Female']

# group the data by season and count the number of pitches in each season
season_counts = female_pitches.groupby('Season Number')['Pitch Number'].count()

# find the season with the highest count
max_season = season_counts.idxmax()
max_count = season_counts[max_season]

# print the result
print(f"Season {max_season} had the most female pitches with {max_count} pitches.")
```

```python
#Which season had the least female pitches?

# filter the data for pitches thrown by female pitchers
female_pitches = data[data['Pitchers Gender'] == 'Female']

# group the data by season and count the number of pitches in each season
season_counts = female_pitches.groupby('Season Number')['Pitch Number'].count()

# find the season with the fewest female pitches
min_season = season_counts.idxmin()
min_count = season_counts[min_season]

# print the result
print(f"Season {min_season} had the fewest female pitches with {min_count} pitches.")
```

```python
#Industry wise accepted deals top 10 in frequency


# group the data by business description and got deal, and count the Pitches column for each group
bd_deals = data.groupby(['Industry', 'Got Deal'])['Pitch Number'].count().reset_index()

# filter for rows where the deal was accepted
bd_accepted = bd_deals[bd_deals['Got Deal'] == 1]

# sort by number of pitches and select top 10
bd_top10 = bd_accepted.sort_values(by='Pitch Number', ascending=False).head(10)

# create a horizontal bar chart of the top 10 business descriptions by number of pitches where the deal was accepted
sns.barplot(x='Pitch Number', y='Industry', data=bd_top10, color='blue')

# set the title and axis labels
plt.title("Top 10 Business Descriptions by Number of Accepted Pitches")
plt.xlabel("Number of Pitches")
plt.ylabel("Business Description")

# show the plot
plt.show()



#Food and Beverage industry, Lifestyle/Home, Fashion/Beaty had the most accepted deals
```

```python
# group the data by industry and got deal, and count the Pitches column for each group
industry_deals = data.groupby(['Industry', 'Got Deal'])['Pitch Number'].count().reset_index()

# pivot the data to create a table of the number of pitches and percentage of deals accepted for each industry
industry_pivot = industry_deals.pivot(index='Industry', columns='Got Deal', values='Pitch Number').reset_index()
industry_pivot['Percentage Accepted'] = industry_pivot[1] / (industry_pivot[1] + industry_pivot[0]) * 100

# sort the data by percentage of deals accepted in descending order
industry_pivot = industry_pivot.sort_values('Percentage Accepted', ascending=False)

# create a bar chart of the percentage of deals accepted for each industry
sns.barplot(x='Percentage Accepted', y='Industry', data=industry_pivot)

# set the title and axis labels
plt.title("Percentage of Deals Accepted by Industry")
plt.xlabel("Percentage of Deals Accepted")
plt.ylabel("Industry")

# show the plot
plt.show()


#According to automotive deals that came most deals were accepted for Automotive about 80%

```

```python
#All about royalty deals

royalty_deals = data[data['Royalty Deal'] == 1]

investments = {
    'Barbara Corcoran': royalty_deals['Barbara Corcoran Investment Amount'].sum(),
    'Mark Cuban': royalty_deals['Mark Cuban Investment Amount'].sum(),
    'Lori Greiner': royalty_deals['Lori Greiner Investment Amount'].sum(),
    'Robert Herjavec': royalty_deals['Robert Herjavec Investment Amount'].sum(),
    'Daymond John': royalty_deals['Daymond John Investment Amount'].sum(),
    'Kevin O\'Leary': royalty_deals['Kevin O Leary Investment Amount'].sum(),
    'Guest': royalty_deals['Guest Investment Amount'].sum()
}

most_royalties = max(investments, key=investments.get)
print(most_royalties)
```

```python
#All about loans

loan_deals = data[data['Loan'] == 1]

investments = {
    'Barbara Corcoran': loan_deals['Barbara Corcoran Investment Amount'].sum(),
    'Mark Cuban': loan_deals['Mark Cuban Investment Amount'].sum(),
    'Lori Greiner': loan_deals['Lori Greiner Investment Amount'].sum(),
    'Robert Herjavec': loan_deals['Robert Herjavec Investment Amount'].sum(),
    'Daymond John': loan_deals['Daymond John Investment Amount'].sum(),
    'Kevin O\'Leary': loan_deals['Kevin O Leary Investment Amount'].sum(),
    'Guest': loan_deals['Guest Investment Amount'].sum()
}

most_loans = max(investments, key=investments.get)
print(most_loans)
```

```python
#The one where the Valuation Requestion was the actual deal valuation

data['Same Valuation'] = (data['Valuation Requested'] == data['Deal Valuation'])

# count the number of deals where the valuation requested is the same as the deal valuation
num_same_valuation = data['Same Valuation'].sum()

print(f"The valuation was the same as the deal valuation for {num_same_valuation} deals.")
```

```python
#Who invested more?

# Calculate total investment amount for each shark
shark_investments = data[["Barbara Corcoran Investment Amount", "Mark Cuban Investment Amount",
                        "Lori Greiner Investment Amount", "Robert Herjavec Investment Amount",
                        "Daymond John Investment Amount", "Kevin O Leary Investment Amount",
                        "Guest Investment Amount"]].sum()

# Sort the values in descending order
shark_investments = shark_investments.sort_values(ascending=False)

# Plot the bar chart
plt.bar(shark_investments.index, shark_investments.values)
plt.title("Total Investment Amount of Each Shark")
plt.xlabel("Shark Name")
plt.ylabel("Investment Amount (in millions)")
plt.xticks(rotation=90)
plt.show()
```
